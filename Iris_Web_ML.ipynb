# Prediction sur les fleurs d'Iris est ce  verginica ou setosa ou versicolor

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
#sklearn est quentient datasets par défaut de Iris 
```

```python
iris=datasets.load_iris()
```

```python
iris.data
# ona 4 input 
#sont des catégories sont les demension pour pour X fleurs  
#width length
```

```python
iris.target 
#On remarque que on 3 3 catégories pour 
```

```python
#On créer notre Objet Model 
clf=RandomForestClassifier()
clf.fit(iris.data,iris.target)
#le syntaxe Génerale de La commande fit 
#model.fit(X_train, y_train)
#pour entraîner un modèle sur un ensemble de données.
```

```python
prediction=clf.predict([[5,2,4,3]])
print(prediction)
```

```python
iris.target_names
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Utiliser Seaborn pour créer un pairplot
sns.pairplot(iris_df, hue='species', diag_kind='hist')
plt.show()
```
